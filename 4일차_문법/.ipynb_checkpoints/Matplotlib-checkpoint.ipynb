# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4,5,6,5,4,3,2,1])
plt.show()
```

```python
## sin 함수 그리기
x = np.arange(0, 12, 0.01)  # 0 ~ 12까지 0.01 간격
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

## Title 설정

```python
plt.plot(x,y)
plt.title('sin graph')
plt.show()
```

## Grid 설정

```python
plt.plot(x,y)
plt.grid()
plt.show()
```

## X, Y 라벨

```python
plt.plot(x,y)
plt.xlabel('time')
plt.ylabel('Amplitude') # 진폭
plt.show()
```

##  그래프 두 개 그리기

```python
plt.plot(x,y)
plt.plot(x, y-1)
plt.plot(x, y-3)
plt.xlabel('time')
plt.ylabel('Amplitude') 
plt.show()
```

## 범례(legend)

```python
plt.plot(x, np.sin(x), label = 'sin')
plt.plot(x, np.cos(x), label = 'cos')
plt.grid()
# plt.legend()
plt.legend(loc = (1.05, 0.05)) # loc=Location String 줄임말로 범례 위치 이동 가능
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('sin & cos')
plt.show()
```

### 범위제한(limit)

```python
plt.plot(x,np.sin(x))
plt.plot(x, np.cos(x))
plt.grid()
plt.ylim(-1.2, 1.2)
plt.xlim(0, np.pi)
plt.show()
```

## 산점도 그래프(scatter)
- 좌표를 이용해 좌표상의 점들을 표시함으로써 두 개 변수 간의 관계를 나타내는 그래프
- x축과 y축의 변수 연관성을 보기 위해 자주 사용

```python
x1=np.random.rand(100)
y1=np.random.rand(100)
x2=np.random.rand(100)
y2=np.random.rand(100)

plt.scatter(x1, y1)
plt.scatter(x2, y2)
```

```python
x1=np.random.rand(100) - 1.0   #x축 범위를 -1.0부터 시작
y1=np.random.rand(100)
x2=np.random.rand(100)
y2=np.random.rand(100)

plt.scatter(x1, y1, marker="+")
plt.scatter(x2, y2, marker="x")
```

```python
import pandas as pd
from matplotlib import pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/datagy/data/main/KNN_data.csv')
df.head()
df.plot()
```

```python
# Adding multiple data labels
df1 = df[df['Label'] == 'Small']
df2 = df[df['Label'] == 'Medium']
df3 = df[df['Label'] == 'Large']

ax = df1.plot(x='x', y='y', kind='scatter', c='r', label='Small')
df2.plot(x='x', y='y', kind='scatter', ax=ax, c='g', label='Medium')  #ax = ax차트를 기준으로 ax 차트상에 df2 차트 표현
df3.plot(x='x', y='y', kind='scatter', ax=ax, c='b', label='Large')
plt.show()
```

## Boxplot(상자도표)

```python
s1 = np.random.normal(loc=0, scale=1, size=1000) # loc는 평균, scale는 표준편차
s2 = np.random.normal(loc=5, scale=0.5, size=1000)
s3 = np.random.normal(loc=10, scale=2, size=1000)
```

```python
plt.plot(s1, label='s1')
plt.plot(s2, label='s2')
plt.plot(s3, label='s3')
plt.legend()
plt.show()
```

```python
plt.boxplot((s1,s2,s3))
plt.grid()
plt.show()
```

## 원그래프(Pie)

```python
import matplotlib.pyplot as plt
import matplotlib

#한글 폰트를 적용하기 위해 필요
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windows
matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
values = [35, 25, 15, 13, 10, 2] # 과목 선호도 조사
# values = [1,1,1,1,1,1]
labels = ['Math', 'Art', 'Korean', 'English', 'Physics', 'Music']

plt.pie(values, labels=labels, autopct='%.1f') # 소수점 첫 번째자리까지 보여준다.
# plt.pie(values, labels=labels, autopct='%.1f%%')
# plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90) # 90도에서 시작
# plt.pie(values, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False) # 시계방향으로
plt.show()
```

```python
values = [35, 25, 15, 13, 10, 2] # 과목 선호도 조사
labels = ['Math', 'Art', 'Korean', 'English', 'Physics', 'Music']
# explode = [0.2, 0.1, 0, 0, 0, 0]
explode = [0.05] * 6  # 라벨간 이격 사이즈

plt.pie(values, labels=labels, explode=explode)
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.legend(loc=(1.2, 0.3))  # 범례 설정. x축, y축 좌표 넣음
plt.show()
```

```python
plt.pie(values, labels=labels, explode=explode)
plt.title("Subjects")
plt.legend(loc=(1.2, 0.3), title='과목 선호도')  #한글폰트 적용하기 위해 rcPanams 함수 호출해야 함(위에서 함)
plt.show()
```

## 막대그래프

```python
labels = ['korea', 'usa', 'china']
values = [100, 130, 150]
colors = ['r', 'g', 'b']

# plt.bar(labels, values)
plt.bar(labels, values, color=colors, alpha=0.5)
```

```python
plt.bar(labels, values)
plt.ylim(90,160)  # y축의 조정으로 kor와 china간 갭이 6배 이상으로 커보이는 착시현상 유도
```

```python
plt.bar(labels, values, width=0.5)
plt.xticks(rotation=45) # x축의 이름 데이터 각도를 45도로 설정
plt.yticks(rotation=45) # y축 이름 데이터 각도 45도 회전
```

```python
ticks = ['1번', '2번', '3번']
plt.bar(labels, values) #x,y축 설정
plt.xticks(labels, ticks, rotation=90) #x축 이름 변경
```
