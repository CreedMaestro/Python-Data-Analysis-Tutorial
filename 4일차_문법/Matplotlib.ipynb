# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,3,4,5,6,5,4,3,2,1])
plt.show()
```

```python
## sin 함수 그리기
x = np.arange(0, 12, 0.01)  # 0 ~ 12까지 0.01 간격
y = np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

## Title 설정

```python
plt.plot(x,y)
plt.title('sin graph')
plt.show()
```

## Grid 설정

```python
plt.plot(x,y)
plt.grid()
plt.show()
```

## X, Y 라벨

```python
plt.plot(x,y)
plt.xlabel('time')
plt.ylabel('Amplitude') # 진폭
plt.show()
```

##  그래프 두 개 그리기

```python
plt.plot(x,y)
plt.plot(x, y-1)
plt.plot(x, y-3)
plt.xlabel('time')
plt.ylabel('Amplitude') 
plt.show()
```

## 범례(legend)

```python
plt.plot(x, np.sin(x), label = 'sin')
plt.plot(x, np.cos(x), label = 'cos')
plt.grid()
# plt.legend()
plt.legend(loc = (1.05, 0.05)) # loc=Location String 줄임말로 범례 위치 이동 가능
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('sin & cos')
plt.show()
```

### 범위제한(limit)

```python
plt.plot(x,np.sin(x))
plt.plot(y, np.cos(x))
plt.grid()
plt.ylim(~1.2, 1.2)
plt.xlim(0, np.pi)
plt.show()
```
